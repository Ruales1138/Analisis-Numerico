|x   |   17 |   19 |   20 |   22 |   23 |   25 |   31 |   32 |   33 |   36  |   37   |   38   |   39|    41|
|----|------|------|------|------|------|------|------|------|------|-------|--------|--------|-----|------|
|y   | 19   | 25   | 32   | 51   | 57   | 71   | 141  | 123  | 187  | 192   | 205    | 252    | 248 | 294  |

1. Hacer un diagrama de  los datos observados. 
2. Construir un polinomio de máximo grado  y  úselo para aproximar  P(18)  y  P(28)  y que puede concluir o interpretar
3. Construir un modelo lineal usando minimos cuadrados y aproxime   P(18)  y  P(28)  y que puede concluir o interpretar
4. Qué puede decir de los dos modelos?


```python
# 1 

import numpy as np
import matplotlib.pyplot as plt
import modelos as md

x_data = np.array([17 , 19 , 20 , 22 , 23 , 25 , 31 , 32 , 33 , 36, 37 , 38 , 39,41])
y_data = np.array([19, 25, 32, 51, 57, 71, 141  , 123  , 187  , 192, 205 , 252 , 248 , 294])
x_values = np.linspace(min(x_data), max(x_data), 10)
plt.plot(x_data, y_data, 'gd', label='Datos observados')
plt.plot(x_values, P(x_values), 'k', label='MC')
plt.legend()
plt.grid()
```

```python
# 2

import sympy as sp

x = sp.symbols('x')
p_maximo = md.lagrange(x_data, y_data)
p_lagrange = sp.lambdify(x, p_maximo)
print(p_lagrange(18))
print(p_lagrange(28))
```

```python
# 3

m, b = md.minimos_cuadrados(x_data, y_data)
P = lambda x: m*x+b
print(P(18))
print(P(28))
```

### 4. ¿Qué puede decir de los dos modelos?

El polinomio de grado máximo interpola exactamente todos los datos dados, pero suele ser muy sensible y puede oscilar entre puntos (especialmente para estimaciones fuera o entre datos lejanos), por lo que sus predicciones pueden ser inestables. En cambio, el modelo lineal por mínimos cuadrados no pasa por todos los puntos, pero captura la tendencia general de crecimiento de manera más estable y robusta. Por eso, para interpretación global y predicción práctica suele preferirse mínimos cuadrados, mientras que el polinomio de grado máximo es útil cuando se requiere ajuste exacto en los nodos.
