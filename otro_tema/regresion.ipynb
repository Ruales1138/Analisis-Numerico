# Regresion por minimos cuadrados

```python
import numpy as np
import matplotlib.pyplot as plt

x_h = np.array([0 ,1.525 ,3.050 ,4.575 ,6.10 ,7.625 ,9.150], float)
y_p = np.array([1, 0.8617, 0.7385, 0.6292, 0.5238, 0.4481, 0.3741], float)

def minimos_cuadrados(x_data, y_data):
    n = len(x_data)
    Sx = sum(x_data)
    Sy = sum(y_data)
    Syx = sum(x_data * y_data)
    Sx2 = sum(x_data ** 2)
    m = (Sy * Sx - n * Syx) / (Sx**2-n*Sx2)
    b = (Sx*Syx-Sy*Sx2) / (Sx**2-n*Sx2)
    return m, b

m, b = minimos_cuadrados(x_h, y_p)
P = lambda x: m*x+b
x_values = np.linspace(min(x_h), max(x_h), 10)
plt.plot(x_h, y_p, 'gd', label='Datos observados')
plt.plot(x_values, P(x_values), 'k', label='MC')
plt.legend()
plt.grid()
```
