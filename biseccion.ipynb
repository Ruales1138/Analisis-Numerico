```python
import numpy as np

def Biseccion(f, a, b, tol):
    if (f(a)*f(b)>0):
        print('No hay nada')
    else:
        while (abs(b-a) > tol):
            p = (a+b)/2
            if f(a)*f(p) < 0:
                b=p
            else:
                a=p
        print(f'La solucion de la funcion: {p}')
    return
```

```python
f = lambda x: x**2-4
Biseccion(f, 1, 3, 1e-16)
```

```python
f = lambda x: 3**(-x)-x
Biseccion(f, 0, 3, 1e-8)
```
