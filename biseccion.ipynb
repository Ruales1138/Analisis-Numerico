```python
import numpy as np

def Biseccion(f, a, b, tol):
    if (f(a)*f(b)>0):
        print('No hay nada')
    else:
        while (abs(b-a) > tol):
            p = (a+b)/2
            if f(a)*f(p) < 0:
                b=p
            else:
                a=p
        print(f'La solucion de la funcion: {p}')
    return
```

```python
f = lambda x: x**2-4
Biseccion(f, 1, 3, 1e-16)
```

```python
f = lambda x: 3**(-x)-x
Biseccion(f, 0, 3, 1e-8)
```

Al final de $t$ segundos, la posición del objeto está dada por

suponga que la partícula se desplazo 1.7 pies en 1 segundo. Encuentre con una exactitud de $10^{-5}$, la tasa $\omega$ a la que $\theta$ cambia. Suponga que $g=32,17pies/s^{2}$}

```python
# Recuerde que se raliza el grafico se la fucion de ceros para estimar un intervalo inicial de busquda.

import matplotlib.pyplot as plt

g = 32.17
x = 1.7
w = np.linspace(-5, -1e-3, 100)
omega = lambda w: -g / (2*w**2) * ((np.exp(w) - np.exp(-w)) / 2 - np.sin(w)) - x
plt.plot(w, omega(w))
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('w')
plt.ylabel('f(w)')
plt.title('Plan inclinado')
plt.grid()
plt.show()
```

```python
import numpy as np

def posicion_falsa(f, a, b, tol):
    if f(a)*f(b) > 0:
        print('No hay nada')
    else:
        p = b-f(b)*(a-b) / (f(a)-f(b))
        while abs(b-a) > tol:
            p = b-f(b)*(a-b)/(f(a)-f(b))
            if f(a)*f(p) < 0:
                b=p
            else:
                a=p
        print(f'La solucion de la funcion: {p}')
    return
```

```python
fpro = lambda x: x**10-1
posicion_falsa(fpro, 0, 3, 1e-6)
```
