```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: (x+np.sqrt(x)) * (20-x+np.sqrt(20-x)) - 155.55
ux = np.linspace(-20, 20, 100)
plt.plot(ux, f(ux), 'b')
plt.axhline(0, linestyle='--', color='red')
plt.grid()
plt.show()
```
