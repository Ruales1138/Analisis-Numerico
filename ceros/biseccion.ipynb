## Teorema de Bolzano

Sea una función $f(x)$ continua definida en un intervalo $[a, b]$, entonces si se cumple que $f(a)f(b) < 0$ (con signos opuestos), existe un punto $c \in (a, b)$ tal que $f(c) = 0$.

---

### Notas adicionales para tu notebook:

Para que el teorema sea válido, se deben cumplir dos condiciones principales:

1. **Continuidad:** La función no debe tener saltos o interrupciones en el intervalo cerrado $[a, b]$.
2. **Cambio de signo:** El valor de la función en los extremos del intervalo debe tener signos distintos (uno positivo y otro negativo).



### Aplicación: Método de Bisección

El **método de bisección** es un procedimiento iterativo que utiliza el Teorema de Bolzano para encontrar raíces de una función.

* **Fundamento:** Se basa en realizar divisiones sucesivas del intervalo inicial $[a, b]$.
* **Procedimiento:** En cada división, se localiza el subintervalo que contiene la raíz aplicando nuevamente el teorema de Bolzano (verificando el cambio de signo).
* **Garantía:** Al ser una función continua con signos opuestos en los extremos, se asegura la existencia de al menos un número $p \in (a, b)$ tal que $f(p) = 0$.


---

### Método de Bisección e Iteraciones

El **método de bisección** utiliza divisiones sucesivas del intervalo $[a, b]$ para localizar la raíz de una función continua donde $f(a)$ y $f(b)$ tienen signos opuestos.

#### 1. Consideraciones sobre Cifras Significativas

Si se trabaja con una **tolerancia** de $10^{-t}$ (donde $t \in \mathbb{N}$), se exige trabajar como mínimo con **$t+1$ cifras significativas**.

#### 2. Cálculo del número de iteraciones ($n$)

Para obtener una tolerancia específica $\varepsilon$, tal que el error absoluto $|\alpha - c_c| < \varepsilon$, el número de iteraciones necesarias se calcula mediante la parte entera de la siguiente fórmula:

$$n = \left\lfloor \frac{\log \left( \frac{b-a}{\varepsilon} \right)}{\log(2)} \right\rfloor$$

> **Nota:** Donde $\lfloor n \rfloor$ representa la función piso o parte entera del resultado.


```python
import numpy as np

def Biseccion(f, a, b, tol):
    # 1. Verificación del Teorema de Bolzano:
    # Si f(a)*f(b) > 0, los signos son iguales y no se garantiza una raíz en el intervalo
    if f(a) * f(b) > 0:
        print('No se cumple el teorema de Bolzano: no se garantiza una raíz en este intervalo.')
    else:
        # 2. Proceso iterativo de divisiones sucesivas del subintervalo
        # El ciclo continúa mientras el ancho del intervalo sea mayor a la tolerancia (ε)
        while abs(b - a) > tol:
            # Calcular el punto medio (p) del intervalo actual
            p = (a + b) / 2      
            # 3. Localizar el subintervalo que contiene la raíz:
            # Si f(a) y f(p) tienen signos opuestos, la raíz está en [a, p]
            if f(a) * f(p) < 0:
                b = p  # El nuevo extremo derecho es el punto medio
            else:
                # De lo contrario, la raíz está en el subintervalo [p, b]
                a = p  # El nuevo extremo izquierdo es el punto medio
        # Al finalizar, 'p' contiene la aproximación con el error < tol deseado
        print(f'La solución aproximada de la función es: {p}')
    return

# Corregir
```

```python
f = lambda x: x**2-4
Biseccion(f, 1, 3, 1e-8)
```

```python
f = lambda x: 3**(-x)-x
Biseccion(f, 0, 1, 1e-8)
```
