# Ejercicio 1
Calcular área y perímetro de círculo, y volumen de
esfera.

El perimetro de un circulo es $P=2\pi r$

El area de un circulo es $A=\pi r^2$

El volumen de una esfera es $V=\frac{4}{3}\pi r^3$


```python
import numpy as np

perimetro = lambda r: 2*np.pi*r

print(f'El perimetro de un circulo de radio 5 es {perimetro(5)}')
```

# Ejercicio 2

Calcular la suma de

$$S=1+\frac{1}{2^2}+\frac{1}{3^2}+\cdots+\frac{1}{n^2}$$

```python
def suma(n):
    s = 0
    for i in range(1, n+1):
        s += 1/(i**2)
    return s

print(f'La suma da como resultado {suma(20)}')
```

## Ejercicio 3

Se desea graficar la función:

$$
f(x) = e^{-x}\sin(2\pi x)
$$

en el intervalo:

$$
x \in [0,4]
$$


```python
import matplotlib.pyplot as plt

x = np.arange(0, 4, 1) # args = Punto inicial, punto final y espaciamiento
xx = np.linspace(0, 4, 1000) # args = Punto inicial, punto final y cantidad de puntos

y = np.exp(-xx)*np.sin(2*np.pi*xx)
plt.plot(xx, y)
#plt.ylim(-0.2, 0.2)
```

# Ejercicio 4
Crea un programa que gestione las calificaciones de un
curso usando funciones. El sistema debe:

a. Registrar notas de estudiantes en diferentes evaluaciones.

b. Calcular promedios de cada estudiante.

c Determinar aprobados y reprobados

d. Generar reportes estadísticos


```python
students = {
    "Ana García": [85, 92, 78, 88],
    "Carlos López": [72, 68, 75, 70],
    "María Rodríguez": [90, 95, 92, 94],
    "Juan Martínez": [55, 60, 58, 52],
    "Laura Fernández": [82, 79, 85, 80],
    "Pedro Sánchez": [45, 50, 48, 42],
    "Sofía Gómez": [95, 98, 96, 97],
    "David Ruiz": [65, 62, 68, 64]
}

def calculate_average(students):
    averages = {}
    for key, value in students.items():
        averages[key] = sum(value)/len(value)
    return averages

def determinate_passed(students):
    averages = calculate_average(students)
    passed = {}
    for key, value in averages.items():
        if value >= 60:
            passed[key] = 'Passed'
        else:
            passed[key] = 'Failed'
    return passed

def generate_reports(students):
    averages = calculate_average(students)
    passed = determinate_passed(students)
    total_average = sum(averages.values())/len(averages)
    max_average = max(averages.values())
    min_average = min(averages.values())
    count_passed = list(passed.values()).count('Passed')
    count_failed = list(passed.values()).count('Failed')
    print(f'\nPromedio general: {total_average}')
    print(f'Nota más alta: {max_average}')
    print(f'Nota más baja: {min_average}')
    print(f'Aprobados: {count_passed}')
    print(f'Reprobado: {count_failed}')

        
print(calculate_average(students))
print(determinate_passed(students))
generate_reports(students)
```
