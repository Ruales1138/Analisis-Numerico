# Ejercicio 1
Calcular área y perímetro de círculo, y volumen de
esfera.

El perimetro de un circulo es $P=2\pi r$

El area de un circulo es $A=\pi r^2$

El volumen de una esfera es $V=\frac{4}{3}\pi r^3$


```python
import numpy as np

perimetro = lambda r: 2*np.pi*r

print(f'El perimetro de un circulo de radio 5 es {perimetro(5)}')
```

# Ejercicio 2

Calcular la suma de

$$S=1+\frac{1}{2^2}+\frac{1}{3^2}+\cdots+\frac{1}{n^2}$$

```python
def suma(n):
    s = 0
    for i in range(1, n+1):
        s += 1/(i**2)
    return s

print(f'La suma da como resultado {suma(20)}')
```
