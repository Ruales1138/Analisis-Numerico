### Error Absoluto
Proporciona la diferencia directa entre el valor aproximado y el valor exacto

### Error Relativo
Expresa el error como una fracción del valor exacto, ofreciendo una perspectiva proporcional.

### Error Porcentual
Muestra el error como un porcentaje del valor exacto, facilitando la interpretación.

```python
def error_absoluto(v_e, v_a):
    return abs(v_e - v_a)

def error_absoluto_porcentual(v_e, v_a):
    err_a = error_absoluto(v_e, v_a)
    return f'{err_a*100}%'

def error_relativo(v_e, v_a):
    return abs(v_e - v_a)/abs(v_e)

def error_relativo_porcentual(v_e, v_a):
    err_r = error_relativo(v_e, v_a)
    return f'{int(err_r*100)}%'

print(error_absoluto(0.000034, 0.000054))
print(error_absoluto_porcentual(0.000034, 0.000054))
print(error_relativo(0.000034, 0.000054))
print(error_relativo_porcentual(0.000034, 0.000054))
    
```
